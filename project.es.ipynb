#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python

cinturon_sur.loc[:, 'precio_normalizado'] = cinturon_sur.groupby('level5')['price'].transform(lambda x: (x - x.min()) / (x.max() - x.min()))

# gráfico
plt.figure(figsize=(10, 6))
poblaciones = cinturon_sur['level5'].unique()

for poblacion in poblaciones:
    subset = cinturon_sur[cinturon_sur['level5'] == poblacion]
    plt.hist(subset['precio_normalizado'], bins='auto', alpha=0.5, label=poblacion, edgecolor='black')

plt.title('Precios Normalizados por Población')
plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.legend(title='Poblaciones')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

Pista: Crea una nueva columna llamada `pps` (price per square en inglés) y luego analiza los valores

```python

# Columna 'pps' (precio por metro cuadrado)
cinturon_sur.loc[:, 'pps'] = cinturon_sur['price'] / cinturon_sur['surface']

# Datos por municipio
getafe = cinturon_sur[cinturon_sur['level5'] == 'Getafe']
alcorcon = cinturon_sur[cinturon_sur['level5'] == 'Alcorcón']

# Estadísticas para cada municipio
pps_mean_getafe = getafe['pps'].mean()
pps_mean_alcorcon = alcorcon['pps'].mean()
pps_var_getafe = getafe['pps'].var()
pps_var_alcorcon = alcorcon['pps'].var()


print(f"En Getafe:")
print(f"  Media del precio por metro cuadrado: {pps_mean_getafe:.2f} USD/m²")
print(f"  Varianza del precio por metro cuadrado: {pps_var_getafe:.2f}")

print(f"\nEn Alcorcón:")
print(f"  Media del precio por metro cuadrado: {pps_mean_alcorcon:.2f} USD/m²")
print(f"  Varianza del precio por metro cuadrado: {pps_var_alcorcon:.2f}")
```

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆) 
Pista: Haz un diagrama de dispersión de cada población usando subgráficos (subplots).

```python
## Aquí: traza la coordenadas de los estados

from ipyleaflet import Marker
datos = {
    'latitud': [40.35, 40.36, 40.37],
    'longitud': [-3.8, -3.81, -3.79],
    'población': ['Madrid Sur', 'Madrid Sur', 'Getafe']
}


## PON TU CÓDIGO AQUÍ:

# Transformar las coordenadas en un diccionario de Python
coordenadas_dict = {}
for i in range(len(datos['latitud'])):
    poblacion = datos['población'][i]
    lat = datos['latitud'][i]
    lon = datos['longitud'][i]

    if poblacion not in coordenadas_dict:
        coordenadas_dict[poblacion] = []
    coordenadas_dict[poblacion].append((lat, lon))

# IMPORTANTE!!!: No entiendo el porque pero esta ventana de codigo me ha dado problemas con la anterior por eso el import y el Map los he colocado en esta ventana.
map = Map(center=(40.35, -3.8), zoom=11.5, min_zoom=1, max_zoom=20,
          basemap=basemaps.OpenStreetMap.Mapnik)

# Agregar marcadores al mapa según el diccionario
for poblacion, coordenadas in coordenadas_dict.items():
    for lat, lon in coordenadas:
        marker = Marker(location=(lat, lon), title=poblacion)
        map.add_layer(marker)

#
map
```
